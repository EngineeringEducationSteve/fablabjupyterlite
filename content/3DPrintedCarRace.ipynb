# 3D Printed Car Race
We have several cars that we can inspect so we can learn about possibilities for another car.



```p5js
let i; 

var car1Distances = [];
var car2Distances = [];
var car3Distances = [];

function setup() {
    createCanvas(800, 400);
    background(200);
    i = 0;
}

function drawCar1(x, y) {
    fill(0,0,0); //tire 1
    rect(x+10, y+65, 20, 15, 5);
    fill(0,0,0); //tire 2
    rect(x+110, y+65, 20, 15, 5);
    fill(0,0,0); //tire 3
    rect(x+10, y-10, 20, 15, 5);
    fill(0,0,0); //tire 4
    rect(x+110, y-10, 20, 15, 5);
    fill(20,50,200);//3D Printed Body
    rect(x, y, 150, 70, 20);
    fill(20,150,10);//Raspberry Pi Pico
    rect(x+3, y+25, 50, 20);
    fill(255, 255, 255);//Battery
    rect(x+65, y+2, 25, 65);
    car1Distances.push(x);
}

function drawCar2(x, y) {
    fill(0,0,0); //tire 1
    rect(x+10, y+65, 20, 15, 5);
    fill(0,0,0); //tire 2
    rect(x+110, y+65, 20, 15, 5);
    fill(0,0,0); //tire 3
    rect(x+10, y-10, 20, 15, 5);
    fill(0,0,0); //tire 4
    rect(x+110, y-10, 20, 15, 5);
    fill(240,200,0);//3D Printed Body
    rect(x, y, 150, 70, 20);
    fill(20,150,10);//Raspberry Pi Pico
    rect(x+3, y+25, 50, 20);
    fill(255, 255, 255);//Battery
    rect(x+65, y+2, 25, 65);
    car2Distances.push(x);
}

function drawCar3(x, y) {
    fill(0,0,0); //tire 1
    rect(x+10, y+65, 20, 15, 5);
    fill(0,0,0); //tire 2
    rect(x+110, y+65, 20, 15, 5);
    fill(0,0,0); //tire 3
    rect(x+10, y-10, 20, 15, 5);
    fill(0,0,0); //tire 4
    rect(x+110, y-10, 20, 15, 5);
    fill(170,20,10);//3D Printed Body
    rect(x, y, 150, 70, 20);
    fill(20,150,10);//Raspberry Pi Pico
    rect(x+3, y+25, 50, 20);
    fill(255, 255, 255);//Battery
    rect(x+65, y+2, 25, 65);
    car3Distances.push(x);
}

function draw() {
    if (i < 300) {
        i++;
    }
    clear();
    drawCar1(30 + i, 50);
    drawCar2(30 + (i * 1.2), 150);
    drawCar3(30 + (i * 1.12), 260);
}
```
