We can creat a model of this race here and show a simplified version. If you see any errors in the model, you can correct it using p5.js.  "All models are wrong, some models are useful."

```p5js
let i; 

var car1Distances = [];
var car2Distances = [];
var car3Distances = [];

function setup() {
    createCanvas(800, 400);
    background(200);
    i = 0;
}

function drawCar1(x, y) {
    fill(0,0,0); //tire 1
    rect(x+10, y+65, 20, 15, 5);
    fill(0,0,0); //tire 2
    rect(x+110, y+65, 20, 15, 5);
    fill(0,0,0); //tire 3
    rect(x+10, y-10, 20, 15, 5);
    fill(0,0,0); //tire 4
    rect(x+110, y-10, 20, 15, 5);
    fill(20,50,200);//3D Printed Body
    rect(x, y, 150, 70, 20);
    fill(20,150,10);//Raspberry Pi Pico
    rect(x+3, y+25, 50, 20);
    fill(255, 255, 255);//Battery
    rect(x+65, y+2, 25, 65);
    car1Distances.push(x);
}

function drawCar2(x, y) {
    fill(0,0,0); //tire 1
    rect(x+10, y+65, 20, 15, 5);
    fill(0,0,0); //tire 2
    rect(x+110, y+65, 20, 15, 5);
    fill(0,0,0); //tire 3
    rect(x+10, y-10, 20, 15, 5);
    fill(0,0,0); //tire 4
    rect(x+110, y-10, 20, 15, 5);
    fill(240,200,0);//3D Printed Body
    rect(x, y, 150, 70, 20);
    fill(20,150,10);//Raspberry Pi Pico
    rect(x+3, y+25, 50, 20);
    fill(255, 255, 255);//Battery
    rect(x+65, y+2, 25, 65);
    car2Distances.push(x);
}

function drawCar3(x, y) {
    fill(0,0,0); //tire 1
    rect(x+10, y+65, 20, 15, 5);
    fill(0,0,0); //tire 2
    rect(x+110, y+65, 20, 15, 5);
    fill(0,0,0); //tire 3
    rect(x+10, y-10, 20, 15, 5);
    fill(0,0,0); //tire 4
    rect(x+110, y-10, 20, 15, 5);
    fill(170,20,10);//3D Printed Body
    rect(x, y, 150, 70, 20);
    fill(20,150,10);//Raspberry Pi Pico
    rect(x+3, y+25, 50, 20);
    fill(255, 255, 255);//Battery
    rect(x+65, y+2, 25, 65);
    car3Distances.push(x);
}

function draw() {
    if (i < 300) {
        i++;
    }
    clear();
    drawCar1(30 + i, 50);
    drawCar2(30 + (i * 1.2), 150);
    drawCar3(30 + (i * 1.12), 260);
}
```

# The Graphing Car Distance
We can take the speed to plot a graph of the car's displacement.  Below you will see that we are collecting the locations of the model car over time and then we plt it on the screen.

```p5js
let i; 

var car1Distances = [];
var car1Speed = 1;
var car2Distances = [];
var car2Speed = 1.2;
var car3Distances = [];
var car3Speed = 1.12;
var points = [];

for (i = 0; i < 300; i++) {
    plotCar(1, 30 + (i * car1Speed), 50);
    plotCar(2, 30 + (i * car2Speed), 150);
    plotCar(3, 30 + (i * car3Speed), 260);
}

function plotCar(carnumber, x, y) {
    var myCarData;
    if (carnumber == 1) {
        car1Distances.push(x);
    } else if (carnumber==2) {
        car2Distances.push(x);
    } else {
        car3Distances.push(x);
    }
}

function setup() {
    createCanvas(800, 400);
    background(200);
    i = 0;
}

function draw() {
    var y1 = 400;
    car1Distances.forEach(point => {
        fill(20,50,200);
        circle(point, y1, 5);
        y1 -= 1;
    });
    var y2 = 400;
    car2Distances.forEach(point => {
        fill(240,200,0);
        circle(point, y2, 5);
        y2 -= 1;
    });
    var y3 = 400;
    car3Distances.forEach(point => {
        fill(170,20,10);
        circle(point, y3, 5);
        y3 -= 1;
    });
}
print("Car 1's Speed is " + car1Speed.toString());
print("Car 2's Speed is " + car2Speed.toString());
print("Car 3's Speed is " + car3Speed.toString());
```
