```p5js
let burgers, distance, drive_time, price, quality;

function setup() {
    createCanvas(1000, 400);
    burgers = ["In-N-Out (Manteca)", "In-N-Out (Modesto)", "Chick-Fil-A (Manteca)", "Chick-Fil-A (Lathrop)", "Carl's Junior (Ripon)", "The Habit (Manteca)", "The Habit (Modesto)", "Sonic (Modesto)"];
    distance = [11.1, 11.6, 11, 20, 11.5, 9, 11.2, 9.5];
    drive_time = [19, 25, 18, 32, 18, 24, 18, 27];
    price = [3.69, 3.69, 5.79, 5.79, 7.04, 7.49, 7.49, 5.89];
    quality = [10, 10, 9, 9, 8, 5, 5, 2];
    textSize(15);
}

function drawPoint(my_name, my_icon, my_x, my_y, my_size, my_color) {
    textSize(my_size * 15);
    my_color = 50 + round(my_color) * 20;
    fill(color(255 - my_color, my_color, my_color));
    rect(10 * my_x, 350 - 8 * my_y - 20, my_size * 25, my_size * 5);
    fill(color(0, 0, 0));
    text(my_icon, 10 * my_x + 25, 350 - 8 * my_y - 20);
    textSize(my_size * 5 /2);
    text(my_name, 10 * my_x + 5, 345 - 8 * my_y);
}

function draw() {
    clear();
    drawPoint(burgers[0], "üçî", 5 * 1, drive_time[0], price[0], quality[0]);
    drawPoint(burgers[1], "üçî", 5 * 3, drive_time[1], price[1], quality[1]);
    drawPoint(burgers[2], "üêî", 5 * 5, drive_time[2], price[2], quality[2]);
    drawPoint(burgers[3], "üêî", 5 * 7, drive_time[3], price[3], quality[3]);
    drawPoint(burgers[4], "üçî", 5 * 9, drive_time[4], price[4], quality[4]);
    drawPoint(burgers[5], "üçî", 5 * 11, drive_time[5], price[5], quality[5]);
    drawPoint(burgers[6], "üçî", 5 * 13, drive_time[6], price[6], quality[6]);
    drawPoint(burgers[7], "üçî", 5 * 15, drive_time[7], price[7], quality[7]);
}
```

We can use math to define shapes and create images.  In fact there is some math in any image you see on a screen.  The types of math depends on the image. For instance, many things that look like drawings are vector graphics, defined by a function that draws a line.  3D graphics are also, lines, but they are defined with more dimensions.  For instance, consider a function like y = f(x) = x<sup>2</sup>.  A 3D function would look more like z = f(x, y) = x<sup>2</sup> + y<sup>2</sup>. <br>
We can draw 4 simple rectangles with the code below.  Can you write a few lines of code to change them into an "E"?
