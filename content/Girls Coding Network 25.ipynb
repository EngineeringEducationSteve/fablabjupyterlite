### We Can Model Our Neopixel Grid

Neopixels are spots of color.  We can determine the color of each neopixel.  

Lets start with designing a grid and having a function to fill the grid.  A function defined below "grid_fill" will do that for us.

```python
import matplotlib.pyplot as plt
import numpy as np

dark_pixel = (.05, 0.05, 0.05)  # (Red, Green, Blue) Amount from 0 - 1
red = (1, 0, 0)
blue = (0, 0, 1)
green = (0, 1, 0)
bright = (0.9, 0.9, 0.9)

plt.rcParams['lines.markersize'] = 30

def grid_fill(new_color):
    column_start = 0
    column_end = 8
    row_start = 0
    row_end = 4
    while (row_start < row_end):
        column_start = 0
        while (column_start < column_end):
            plt.scatter(column_start, row_start, color=new_color)
            column_start += 1
        row_start += 1

grid_fill(dark_pixel)
plt.show()
```

We can fill the grid with a new color or a color that we have designed. Below, we have filled the grid with red.  You can change it to something like green, or a color that you define.

```python
grid_fill(red)
plt.show()
```

We can also set specific pixel's colors.  Here we are making the whole drid dark, but with some pixel's turned on to make a smile.  Try editing it to make a different pattern.

```python
grid_fill(dark_pixel)
#left eye
plt.scatter(2, 3, color=green)
plt.scatter(2, 2, color=green)
#right eye
plt.scatter(5, 3, color=green)
plt.scatter(5, 2, color=green)
#smile
plt.scatter(0, 1, color=green)
plt.scatter(1, 0, color=green)
plt.scatter(2, 0, color=green)
plt.scatter(3, 0, color=green)
plt.scatter(4, 0, color=green)
plt.scatter(5, 0, color=green)
plt.scatter(6, 0, color=green)
plt.scatter(7, 1, color=green)
plt.show()
```

### Code A Pattern
Here we have a simple pattern.  You can cut and paste this code into your matrix to see how it will look.  Try editing it on your device. (This is the basic program your device started with.)

```python
import time
import board
import neopixel
#from adafruit_msa3xx import MSA311
#from adafruit_apds9960.apds9960 import APDS9960


#i2c = board.I2C()  # uses board.SCL and board.SDA
#msa = MSA311(i2c)
#apds = APDS9960(i2c)

silver = (186, 186, 186)
red = (255, 0, 0)
yellow = (255,255,0)
orange = (255, 120, 0)
green = (0, 255, 0)
blue = (50, 50, 255)
mauve = (192, 34, 197)
white = (255, 255, 255)

#apds.enable_proximity = True

pixels = neopixel.NeoPixel(board.D6, 32, brightness=0.1)
pixels.fill(red)
pixel_num = 0
while pixel_num < 32:
    if (pixel_num % 8 == 0 or pixel_num % 8 == 7):
        pixels[pixel_num] = green
    if (pixel_num % 8 == 1 or pixel_num % 8 == 6):
        pixels[pixel_num] = green
    if (pixel_num % 8 == 2):
        pixels[pixel_num] = blue
    if (pixel_num % 8 == 3):
        pixels[pixel_num] = yellow
    if (pixel_num % 8 == 4):
        pixels[pixel_num] = orange
    if (pixel_num % 8 == 5):
        pixels[pixel_num] = mauve
    pixel_num += 1
pixels.show()
```

### Tilt
This is an example of code for a fabagatchi that changes its color when tilted different ways.  You'll need to make sure you have the accelerometer (MSA311). 

```python
import time
import board
import neopixel
from adafruit_msa3xx import MSA311

i2c = board.I2C()  # uses board.SCL and board.SDA
msa = MSA311(i2c)

pixels = neopixel.NeoPixel(board.D6, 32, brightness=0.1)
pixels[0] = (255, 255,0)
pixels[1] = (255, 0, 0)
pixels[2] = (0, 255, 0)
pixels.show()
loops = 0
count = 0
while True:
    red_amount = int(msa.acceleration[0] * 10) + 120
    green_amount = int(msa.acceleration[1] * 10) + 120
    count = 0
    while count < 32:
        pixels[count] = (red_amount,green_amount,40+6*count)
        count += 1
    pixels.show()
    time.sleep(.1)
    print(str(red_amount) + ", " + str(green_amount))
```

### Proximity
This is an example of code for a fabagatchi that changes its color when something comes near it. You'll need to maek sure that you have the proximity/gesture/color sensor (APDS-9960).

```python
import time
import board
import neopixel
#from adafruit_msa3xx import MSA311
from adafruit_apds9960.apds9960 import APDS9960


i2c = board.I2C()  # uses board.SCL and board.SDA
#msa = MSA311(i2c)
apds = APDS9960(i2c)


apds.enable_proximity = True

pixels = neopixel.NeoPixel(board.D6, 32, brightness=0.1)
pixels[0] = (255, 255,0)
pixels[1] = (255, 0, 0)
pixels[2] = (0, 255, 0)
pixels.show()
loops = 0
count = 0
while True:
    red_amount = apds.proximity
    green_amount = 0
    count = 0
    while count < 32:
        pixels[count] = (red_amount,green_amount,40+6*count)
        count += 1
    pixels.show()
    time.sleep(.1)
    print(str(red_amount) + ", " + str(green_amount))
```
