# P5.Js Refresher

We can use a collection of simple codes in P5.js to remind us of how to program in this language.

We will work with a canvas of 400 x 400.  This image should help us to consider locations.



We'll start with making a simple rectangle to represent the ocean.



```p5js
function setup() {
    createCanvas(400, 400);
    background(200);
    //Water
    fill(100, 125, 255); //We give the amounts of 3 light primaries, red, green, and blue from 0-255
    rect(0, 350, 400, 50);
    //Sky
    fill(150, 175, 250);
    rect(0, 0, 400, 350);
}

function draw() {
}
```

We can move on from here to add a circle (oval) to this image.  We can add an ellpse to represent an island and a circle to represeent the sun.  A circle is defined by the central point and its radius.



```p5js
function setup() {
    createCanvas(400, 400);
    background(200);
    //Water
    fill(100, 125, 255); //We give the amounts of 3 light primaries, red, green, and blue from 0-255
    rect(0, 350, 400, 50);
    //Sky
    fill(150, 175, 250);
    rect(0, 0, 400, 350);
    //Sun
    fill(255, 255, 102);
    circle(300, 50, 125);
    //Island
    fill(204, 204, 179)
    ellipse(250, 350, 200, 50);
}

function draw() {
}
```

```p5js
%show
```

We can add a tree with large palm leaves made of triangles.


Each triangle is defined by its 3 points.

```p5js
function setup() {
    createCanvas(400, 400);
    background(200);
    //Water
    fill(100, 125, 255); //We give the amounts of 3 light primaries, red, green, and blue from 0-255
    rect(0, 350, 400, 50);
    //Sky
    fill(150, 175, 250);
    rect(0, 0, 400, 350);
    //Sun
    fill(255, 255, 102);
    circle(300, 50, 125);
    //Island
    fill(204, 204, 179)
    ellipse(250, 350, 200, 50);
    //Tree Trunk
    fill(153, 102, 0); 
    rotate(QUARTER_PI/3);
    rect(275, 125, 55, 150, 20);
    rotate(-QUARTER_PI/3);
    //Leaves
    fill(0, 153, 0)
    triangle(265, 215, 300, 185, 315, 285);
    triangle(265, 215, 235, 175, 180, 270);
    triangle(270, 210, 250, 165, 170, 260);
    triangle(250, 200, 280, 175, 330, 295);
}

function draw() {
}
```

```p5js
%show
```
