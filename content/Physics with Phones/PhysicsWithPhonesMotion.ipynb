## Making a Graph
I exported the data from the device to a csv.  I'll use matplotlib to show some of the data and to graph it.


```python
import pandas as pd
import matplotlib.pyplot as plt
walking_df = pd.read_csv('/Users/scallahan/Documents/Jupyter/Notebooks/PWPdata/WalkingData.csv', on_bad_lines='skip')
walking_df.head() #first 5 lines of the CSV
```

```python
walking_df.tail() # Last 5 lines of the CSV
```

```python
walking_df.set_index('Time (s)').plot()
plt.show()
```

The graph above shows the x, y, and z acceleration with respect to time.  It also shows the absolute acceleration.  When do you think I walked up the stairs?

## Interpreting Data
We can also do some analysis on the data.   Here we have taken the average of the accerations.


```python
average_z = walking_df["Linear Acceleration x (m/s^2)"].mean()
print("the average z acceleration was " + str(average_z)+".")
average_y = walking_df["Linear Acceleration y (m/s^2)"].mean()
print("the average y acceleration was " + str(average_y)+".")
```
