# Understanding Motion Data with Jupyter Notebook
This is a how we can help students to analyze data with the jupyter notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('microbit.csv')
df.head()
```

We can make a line plot of our data very quickly.  It doesn't take much work.  One line is enough.

```python
lines = df.plot.line()
```

That is a bit messy, maybe we want to focus on just one of these, like the total acceleration.

```python
lines = df.plot.line(y="total acceleration", x="Time (seconds)")
lines.set_xlim([9, 12])
```
