# Motion
We can show how something moving by changing the location each time that the scene is drawn.

```p5js
var x = 50;

function setup () {
    createCanvas(400, 400);
    background(250);
    fill('cyan');
    circle(50, x, 30);
}

function draw() {
    x++;
    clear
    fill('red');
    circle(x, 50, 30);
}
```

```p5js
%show
```

# Bounce
We could also use logic to make rules so the ball moves forward and then background.

```p5js
var x = 50;
var speed = 2;

function setup () {
    createCanvas(400, 400);
    background(250);
    fill('cyan');
    circle(x, 50, 30);
}

function draw() {
    if (x > 400) {
        speed = -2;
    }
    if (x < 0) {
        speed = 2;
    }
    x += speed;
    circle(x, 50, 30);
}
```

```p5js
%show
```

# Reflection
This quick model of reflection shows the light bouncing off of a surface.  Does this fit with your observation of the light in the laser

```p5js
var x = 10;
var y = 10;
var speed = 2;
var mirror_x = 150
var mirror_y = 150

function setup () {
    createCanvas(400, 400);
    background(15); //Dark room
    fill('grey');
    stroke('grey');
    rect(mirror_x, mirror_y - 50, 20, 100); //Mirror Surface
}

function draw() {
    if (y <= mirror_y) {
        x += speed;
        y += speed;
    } else {
        x -= speed;
        y += speed;
    } 
    fill('red');
    stroke('red');
    circle(x, y, 5); //Laser light
}
```

```p5js
%show
```

# Refraction
This simple model shows the light moving through the lense.  Does this model seem correct, given what you have seen?

```p5js
var x = 10;
var y = 10;
var speed = 2;
var lense_x = 150;
var lense_y = 150;
var lense_width = 20;

function setup () {
    createCanvas(400, 400);
    background(15); //Dark room
    fill('white');
    stroke('white');
    rect(lense_x, lense_y - 50, lense_width, 100); //Lense 
}

function draw() {
    if ((x >= lense_x) && (x <= (lense_x + 20))) {
        x += speed / 2;
        y += speed;
    } else {
        x += speed;
        y += speed;
    } 
    fill('red');
    stroke('red');
    circle(x, y, 5); //Laser light
}
```

```p5js
%show
```

# Filtration
This model shows light traveling through the filters.  Does this fit with what you have observed?  How can we improve it to be more like our data.

```p5js
var beam_x = 10;
var beamR_y = 50;
var beamG_y = 150;
var beamB_y = 250;
var speed = 2;
var filter_x = 150;
var filterR_y = 25;
var filterG_y = 125;
var filterB_y = 225;
var filter_width = 10;


function setup () {
    createCanvas(400, 400);
    background(15); //Dark room
    fill('red');
    stroke('red');
    rect(filter_x, filterR_y, filter_width, 75); //red filter
    fill('green');
    stroke('green');
    rect(filter_x, filterG_y, filter_width, 75); //gree filter
    fill('blue');
    stroke('blue');
    rect(filter_x, filterB_y, filter_width, 75); //blue filter
}

function draw() {
    if (beam_x < filter_x) {
        fill('white')
        stroke('white')
    } else {
        fill('red')
        stroke('red')
    }
        circle(beam_x, beamR_y, 5); //Light through red filter
    if (beam_x < filter_x) {
        fill('white')
        stroke('white')
    } else {
        fill('green')
        stroke('green')
    }
        circle(beam_x, beamG_y, 5); //Light through green filter
    if (beam_x < filter_x) {
        fill('white')
        stroke('white')
    } else {
        fill('blue')
        stroke('blue')
    }
        circle(beam_x, beamB_y, 5); //Light through blue filter
    beam_x += speed;
}
```
